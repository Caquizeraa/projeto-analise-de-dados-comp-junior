## Exercício 03
Como parte do seu relatório mensal, você precisa fornecer um resumo estatístico dos preços dos produtos. Utilize o método describe() do Pandas para detalhar a distribuição dos preços, destacando possíveis outliers e sua possível influência no cálculo da média.

```python
# Código referente ao exercício 03
#Describe geral
print('Describe:')
print(items_df['price'].describe(),'\n')

# Disperção
ap = items_df['price'].max()-items_df['price'].min()
iqr = np.array(items_df['price'].describe()[6:7].values - items_df['price'].describe()[4:5].values)[0]
print('Amplitude:',ap)
print('Amplitude interquartilica:',iqr)
print('Desvio Padrão:',round(items_df['price'].std(),2))
print('Variância:',round(items_df['price'].var(),2))

# Plotando disperção
plt.figure(figsize=(12,6))
sns.boxplot(data=items_df,x='price')
plt.title('Análise de dispersão de Preços')
plt.xlabel('Preço (R$)');

# Analisando esses dados, podemos perceber que existem uma alta disperção entre os preços
# Podemos perceber que são valores muito heterogêneos e a existência de muitos outliers
# Alguns motivos que tornam possível perceber isso são:
#  -Grande diferença entre a amplitude e a amplitude interquatilica
#  -Valor significativo de diferença entre a média e a mediana
#  -Valor de desvio padrão maior que a média
#  -O gráfico mostra a existência dos outliers por meio de losangos, e vemos que existem muitos
#  -Valor discrepante entre a média e os dois valores extremos (mínimo e máximo)
# Portanto, é possível perceber que a média é uma medida muito sensível a esses outliers,
# e a tendência central pode ser percebida de outras formas, como a mediana
```

## Exercício 04
A eficiência logística é vital para a satisfação do cliente. Adicione uma coluna ao DataFrame de pedidos que calcule o número de dias entre a data de compra e a data de entrega. Discuta como variações nesse prazo podem impactar a percepção do cliente.

```python
# Código referente ao exercício 04
# Transoformando os valores de Data de entrega e Data de compra para o tipo de dados: datetime
orders_df_clean['order_approved_at'] = pd.to_datetime(orders_df_clean['order_approved_at'])
orders_df_clean['order_delivered_customer_date'] = pd.to_datetime(orders_df_clean['order_delivered_customer_date'])

# Calculando a diferença entre a Data de entrega e a Data de compra
# E depois pegando o valor em dias desse resultado
orders_df_clean['order_deliver_daycount'] = (
    abs((orders_df_clean['order_delivered_customer_date'] - orders_df_clean['order_approved_at']).dt.days)
)

# Analisando esses dados 
print(orders_df_clean['order_deliver_daycount'].describe())

# Percebi que o prazo de entrega médio era de 11 dias
# Também percebi que 75% das entregas tem um tempo de entrega de até 15 dias para a entrega
# Existem entregas com te,po maior que de 15 dias, 
# e esse longo período tempo trás uma impressão negativa sobre a loja!
# Percebemos uma entrega que demorou 208 dias, o que é um problema,
# pois esse prazo não é comum no mercado, e pode ocasionar a perda do cliente.

# Plotando prazos
plt.figure(figsize=(12,6))
sns.boxplot(data=orders_df_clean,x='order_deliver_daycount')
plt.title('Análise de dispersão de Prazos')
plt.xlabel('Tempo (Dias)');
```

## Exercício 05
Visualizações claras podem revelar tendências que antes não eram possíveis de serem capturadas. Crie um gráfico de barras vertical que mostre o volume de vendas mensal por estado. Use essa visualização para identificar quais estados apresentaram maior crescimento no último ano.

## Exercício 08
A precisão nas previsões de entrega é um indicador chave de desempenho logístico. Desenvolva um método para comparar as datas estimadas e reais de entrega, identificando padrões de atrasos ou adiantamentos nas entregas. Faça um gráfico de linhas com o tempo médio de atraso por semana ao longo do tempo. O eixo y deve ser o tempo médio de atraso (em dias) e o eixo x as semanas do ano. 

```python
# Código referente ao exercício 08

# Passando os valores para datetime
df_clean['order_estimated_delivery_date'] = pd.to_datetime(df_clean['order_estimated_delivery_date'])
df_clean['order_delivered_customer_date'] = pd.to_datetime(df_clean['order_delivered_customer_date'])

# Descobrindo qual semana do ano é
df_clean['week'] = df_clean['order_approved_at'].dt.isocalendar().week

# Calculando atraso (valores positivos) ou adiantamento (valores negativos)
df_clean['precision'] = (
     df_clean['order_delivered_customer_date'] - df_clean['order_estimated_delivery_date'] 
).dt.days

# Escolhendo o último ano (o mais recente é 2018)
analise_semanal = df_clean.loc[
        df_clean['year'] == 2018
]

# Agrupando por semana
analise_semanal = df_clean.groupby(by=['week']).mean(numeric_only=True).sort_values(['week'])

# Plotando o gráfico
plt.figure(figsize=(10,5))
plt.plot(analise_semanal.index,analise_semanal['precision'])
plt.title('Média de atraso por Semana em 2018')
plt.xlabel('Semana do ano')
plt.ylabel('Média de atraso');

```

## Exercício 09
Investigue a relação entre o valor do frete e o preço dos produtos. Calcule a correlação entre esses dois fatores e crie um scatter plot para visualizar essa relação, discutindo como isso pode afetar as decisões de frete grátis.

```python
# Código referente ao exercício 09
# Calculando a correlação
pearson = df_clean.corr(numeric_only=True)
spearman = df_clean.corr('spearman', numeric_only=True)

# Separando as colunas de frete e preço
pearson = pearson['price'].loc[
    pearson.index=='freight_value'
]

spearman = spearman['price'].loc[
   spearman.index=='freight_value'
]

# Imprimindo no console
print('Valores de correlação')
print('Correlação de Pearson:',pearson['freight_value'])
print('Correlação de Spearman:',spearman['freight_value'])

# Plotando gráfico
plt.figure( figsize=(10, 6) )
plt.title('Análise de Correlação')
plt.scatter( df_clean['price'], df_clean['freight_value']);
plt.xlabel('Preço')
plt.ylabel('Frete');

# Acredito que existe correlação, porém não muito significativa.
# Pois ambas as medidas podem estar entre 0 a 1, 
# mas não valem nem 50% desse range.
# A decisão de frete grátis deve analisar outros fatores perante a saúde financeira da loja
# para ser tomada.
```
